```python
# Define an acceleration function for the Bando:
a = .5
b = 20.0
s_0 = 10.0
s_star = .5
v_m = 30.0

def f(s,v,delta_v):
    # (TODO) implement accel function for Bando:
    acc = a*(OV(s) - v) + b*(delta_v/(s**2))
    return acc

def OV(s):
    #optimal velocity function:
    v_des = v_m*(np.tanh(s/s_0 - s_star) + np.tanh(s_star))/(1 + np.tanh(s_star))
    return v_des

# Plot the optimal velocity function:
s_vals = np.linspace(0,50,100)
v_vals = OV(s_vals)

pt.figure()
pt.plot(s_vals,v_vals)
pt.xlabel('Equilibrium spacing')
pt.ylabel('Equilibrium speed')
pt.title('Optimal Velocity Model')
pt.show()
# Try changing around the parameters for the given model to
# see how the optimal velocity function changes.
```

```python
# Write a function to perform a simulation:

def simulate_follow(s_0,v_0,v_lead,times):
    v_sim = np.zeros(np.shape(v_lead))
    v_sim[0] = v_0
    s_sim = np.zeros(np.shape(v_lead))
    s_sim[0] = s_0
    num_Steps = int(len(times))
        
    for t in range(1,num_Steps):
        delta_T = times[t] - times[t-1]
        # (TODO) Implement stepping routine to solve:
        delta_v = (v_lead[t-1] - v_sim[t-1])
        acc = f(s_sim[t-1],v_sim[t-1],delta_v)
        v_sim[t] = v_sim[t-1] + delta_T*acc
        s_sim[t] = s_sim[t-1] + delta_T*(v_lead[t-1] - v_sim[t-1])
    return v_sim,s_sim
        

s_0 = 30.0
v_0 = v_lead[0]

v_sim,s_sim = simulate_follow(s_0,v_0,v_lead,times)

pt.figure(figsize=(10, 6))
pt.plot(times,v_lead)
pt.plot(times,v_sim)
pt.xlabel('Time [s]')
pt.ylabel('Speed [m/s]')
pt.legend(['Lead vehicle','Simulated vehicle'])
pt.show()

pt.figure(figsize=(10, 6))
pt.plot(times,s_sim)
pt.xlabel('Time [s]')
pt.ylabel('Spacing [m]')
pt.show()
# Make sure that your results here match the figures in the
# report description. 
```
